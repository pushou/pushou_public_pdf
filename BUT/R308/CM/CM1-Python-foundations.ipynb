## Les références en Python


Une référence est un pointeur (pas de complexité en Python mais pas de magie non plus) vers un objet.
Un changement sur l'objet est visible sur
l'ensemble des références. 


```python
a = [1,2,3,4,5]
b = a
print(b)
a[1] = 6
print(a)
print(b)

```

Attention tant que l'objet initial dispose d'une référence sur lui-même il n'est pas détruit:


```python
a=list('azerty')
print(a)
b=a
del a
print('b='+repr(b))
```

## Les fonctions en Python
Un fonction en Python commence par le mot clef def et supporte des annotations de type permettant de clarifier de ce que fais la fonction.


```python
def helloWorld(nom:):
    """ Cette fonction retourne un string"""
    if nom == 'numéro 6':
        return 'je ne suis pas un numéro !'
    return 'Hello ' + nom

helloWorld('numéro 6')
```

C'est plus clair avec une annotation (typing)

```python
def helloWorld(nom:str)->str:
   """ Cette fonction retourne un string"""
   if nom == 'numéro 6':
       return 'je ne suis pas un numéro !'
   return 'Hello ' + nom

helloWorld('pouchou')
helloWorld('numéro 6')
```

Une annotation n'est pas contraignante:
C'est juste une indication pour le (re)lecteur de votre code. 

Mais il vous est demandé **à partir de maintenant d'annoter vos codes**.

### Argument positionnel et argument nommé


```python
def demoArg(*positionnels:tuple,**nommes:dict)->str:
    print(type(positionnels))
    print(type(nommes))
    return f"{positionnels=},{nommes=}"
 
print(demoArg('un','deux','trois'))
print('#'*80)
print(demoArg('un','deux','trois',a='un',b='deux',c='trois'))
print(demoArg())
```

Un argument nommé peut être positionné par défaut:

```python
def addDeuxEntiers(x:int=1,y:int=2)->int:
    return x + y
addDeuxEntiers()
addDeuxEntiers(x=2.5,y=4)
```

Mais attention à l'ordre!

les arguments nommés se doivent d'être positionnés après les arguments positionnels:

```python
print(result:=demoArg('un',b='deux','deux','trois',a='un',c='trois'))
```

```python
def demoArg(*positionnels:tuple,**nommes:dict)->str:
    print(type(positionnels))
    print(type(nommes))
    return f"{positionnels=},{nommes=}"
 
print(result:=demoArg('un','deux','trois'))
```
